```python
import yfinance as yf
import sweetviz as sv
import pandas as pd
```

```python
ticker_IBOV = ['VALE3', 'PETR4', 'WEGE3', 'ITUB4', 'MGLU3',
               'BBDC4', 'ABEV3', 'BBAS3', 'ELET3', 'RENT3',
               'ITSA4', 'JBSS3', 'SUZB3', 'LREN3', 'EQTL3']
```

```python
ticker_IBOV = [x + '.SA' for x in ticker_IBOV]
ticker_IBOV
```

```python
ticker_IBOV.extend(['^BVSP','BRL=x'])
ticker_IBOV
```

```python
data_inicio = '2012-08-23'
data_fim = '2023-02-01'
```

```python
dados_close = yf.download(ticker_IBOV, start=data_inicio, end=data_fim)['Adj Close']
dados_close
```

```python
visual = sv.analyze(dados_close)
```

```python
visual.show_html("relmf.html")
```
